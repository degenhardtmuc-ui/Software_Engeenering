```python
# Funktion 3: Vergleicht 2 Listen => liefert 1 Liste mit den gemeinsamen Zahlen

def gemeinsame_zahlen_finden(spieler_zahlen, computer_zahlen):
    alle_treffer = []

    index = 0       # Index = 0 => Weil Python Listen ab Position 0 zählt. Wenn ich bei 1 starten würde, würde ich die erste Reihe überspringen.

    while index < len(computer_zahlen): # Hier beginnt eine Schleife über die Computerreihen. 
                                        # len(computer_zahlen) ist hier 3, weil es 3 Computerreihen gibt.
                                        # Die Schleife läuft also solange index < 3 ist. 
                                        # Das passiert:
                                        # bei index = 0: erste Reihe
                                        # bei index = 1: zweite Reihe
                                        # bei index = 2: dritte Reihe
                                        # bei index = 3: Stopp, weil 3 < 3 falsch ist
                                    
        eine_reihe = computer_zahlen[index]  #Hier hole ich mir eine einzelne Computerreihe aus der großen Liste heraus.
                                             # Wenn index = 0 ist, bekomme ich die erste Reihe.
                                             # Wenn index = 1 ist, bekomme ich die zweite Reihe.
                                             # Wenn index = 2 ist, bekomme ich die dritte Reihe.
        
        treffer = []        # Hier erstelle ich für diese eine Reihe eine leere Trefferliste.

        zahl_index = 0   # Hier setze ich einen zweiten Index auf 0.
                         # Dieser Index ist für die Spielerzahlen.
                         # Ich habe also: 
                         # index für die Computerreihen
                         # zahl_index für die Spielerzahlen
                         # Ich brauche zwei Indizes, weil ich zwei Listen durchgehen möchte.

        while zahl_index < len(spieler_zahlen): # Hier beginnt die zweite while-Schleife.
                                                # Diese Schleife geht alle Spielerzahlen durch.
                                                # len(spieler_zahlen) ist 6, weil der Spieler 6 Zahlen eingegeben hat.
                                                # Die Schleife läuft also von zahl_index = 0 bis zahl_index = 5.
                                                # Bei zahl_index = 6 stoppt sie, weil 6 < 6 falsch ist.
          
            zahl = spieler_zahlen[zahl_index] # Hier hole ich eine einzelne Spielerzahl aus der Liste heraus.
                                              # Wenn zahl_index = 0 ist, hole ich die erste Spielerzahl.
                                              # Wenn zahl_index = 1 ist, hole ich die zweite Spielerzahl.
                                              # Und so weiter.

            if zahl in eine_reihe:            # Hier prüfe ich, ob diese Spielerzahl in der aktuellen Computerreihe vorkommt.
                                              # in bedeutet: ist enthalten in.
                                              # Beispiel: Wenn die Spielerzahl 3 ist und die Computerreihe [1, 3, 4, 33, 40, 47] ist, 
                                              #  dann stimmt die Bedingung, weil 3 in der Liste enthalten ist.
                                              # Warum benutze ich hier if?
                                              # Ich benutze if, weil ich prüfen möchte, ob eine Zahl ein Treffer ist.
                
                treffer.append(zahl)          # Wenn die Zahl ein Treffer ist, wird sie in die Trefferliste gespeichert.
                                              # append() hängt die Zahl hinten an die Liste an.

            zahl_index = zahl_index + 1       # Hier erhöhe ich den Spielerzahlen-Index um 1.
                                              # Das bedeutet: Gehe zur nächsten Spielerzahl.
                                              # Das ist bei einer while-Schleife sehr wichtig. 
                                              #  Wenn ich das vergesse, bleibt die Schleife immer bei derselben Zahl stehen und läuft endlos weiter.
                                              # Das kann ich im Kurs sagen:
                                              # „Ich erhöhe den Index, damit die Schleife zur nächsten Zahl geht und nicht hängen bleibt.“

        alle_treffer.append(treffer)

        index = index + 1

    return alle_treffer
```
