```python
def ergebnis_auswerten(alle_treffer):
    index = 0

    while index < len(alle_treffer):
        treffer = alle_treffer[index]

        print("Reihe", index + 1, "Treffer:", treffer)

        if len(treffer) > 3:
            print("Gewonnen!")
        else:
            print("Nicht gewonnen.")

        index = index + 1
```
