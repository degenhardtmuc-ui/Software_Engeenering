```python
def spieler_gibt_zahlen():
    spieler_zahlen = []

    while len(spieler_zahlen) < 6:
        zahl = int(input("Geben Sie eine Zahl zwischen 1 und 49 ein: "))

        if zahl >= 1 and zahl <= 49:
            if zahl not in spieler_zahlen:
                spieler_zahlen.append(zahl)
            else:
                print("Diese Zahl haben Sie schon eingegeben!")
        else:
            print("Die Zahl soll zwischen 1 und 49 sein!")

    return spieler_zahlen
```

```python
spieler_gibt_zahlen()
```
